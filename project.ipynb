Построить график смертности по регионам за 2016 год

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data/Число смертей в регионах Приволжья.csv', delimiter=';', encoding='utf-8')
data
x = data['регион']
y = data['число смертей 2016']
plt.figure(figsize=(25, 10))
plt.bar(x, y, color='skyblue')
plt.xlabel('Регионы', fontsize=20)
plt.ylabel('Число умерших 2016 год', fontsize=20)
plt.title('Статистика смерти по регионам за 2016 год', fontsize=20)
plt.grid(True)
plt.show()
```

Построить столбчатые графики для общей смертности и смертности от ДТП по регионам за 2018 год

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data/Число смертей в регионах Приволжья.csv', delimiter=';', encoding='utf-8')
data
regions = data['регион']
deaths = data['число смертей 2018'] 
road_accidents = ['число смертей от ДТП 2018']
index = np.arange(len(regions))
plt.figure(figsize=(20, 10))
plt.bar(index, deaths, color='b', label='Смерти', alpha=0.7)
plt.bar(index, road_accidents, color='r', label='Смертность от ДТП', alpha=0.7, bottom=deaths)
plt.xlabel('Регионы',fontsize=20)
plt.ylabel('Количество',fontsize=20)
plt.title('Гистограмма с накоплением числа смертей и смертностей от ДТП за  2018 год', fontsize=20)
plt.xticks(index, regions, rotation=45)
plt.legend()
plt.show()
```
